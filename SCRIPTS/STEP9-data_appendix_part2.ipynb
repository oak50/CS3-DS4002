## Finalizing the data appendix by adding graphs for "all_celebs_image_data.csv"

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file
celeb = pd.read_csv("/workspace/DS4002Project3/DATA/image_data/celeb_image_data/all_celebs_image_data.csv")

# Summary statistics
print(celeb.describe())

# Standard deviation for brightness
brightness_sd = np.std(celeb['brightness'], ddof=1)  # ddof=1 for sample standard deviation
print(f"Standard Deviation of Brightness: {brightness_sd:.2f}")

# Histogram of brightness
plt.figure(figsize=(8, 5))
sns.histplot(celeb['brightness'], bins=20, color="grey", edgecolor="black")
plt.title("Histogram of Brightness Variable", fontsize=14)
plt.xlabel("Brightness")
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Separate resolution into width and height
celeb[['width', 'height']] = celeb['resolution'].str.split('x', expand=True).astype(int)

# Standard deviation for width
width_sd = np.std(celeb['width'], ddof=1)
print(f"Standard Deviation of Width: {width_sd:.2f}")

# Histogram of width
plt.figure(figsize=(8, 5))
sns.histplot(celeb['width'], bins=20, color="grey", edgecolor="black")
plt.title("Histogram of Width in Resolution Variable", fontsize=14)
plt.xlabel("Width")
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Standard deviation for height
height_sd = np.std(celeb['height'], ddof=1)
print(f"Standard Deviation of Height: {height_sd:.2f}")

# Histogram of height
plt.figure(figsize=(8, 5))
sns.histplot(celeb['height'], bins=20, color="grey", edgecolor="black")
plt.title("Histogram of Height in Resolution Variable", fontsize=14)
plt.xlabel("Height")
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Standard deviation for colorfulness
colorfulness_sd = np.std(celeb['colorfulness'], ddof=1)
print(f"Standard Deviation of Colorfulness: {colorfulness_sd:.2f}")

# Histogram of colorfulness
plt.figure(figsize=(8, 5))
sns.histplot(celeb['colorfulness'], bins=20, color="grey", edgecolor="black")
plt.title("Histogram of Colorfulness Variable", fontsize=14)
plt.xlabel("Colorfulness")
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python

# Load the dataset
celeb = pd.read_csv("/workspace/DS4002Project3/DATA/image_data/celeb_image_data/all_celebs_image_data.csv")

# Ensure 'resolution' column exists and handle missing/invalid data
if 'resolution' not in celeb.columns:
    raise ValueError("The column 'resolution' is missing from the dataset.")
celeb['resolution'] = celeb['resolution'].fillna('0x0')  # Fill missing values
valid_resolutions = celeb['resolution'].str.contains(r'^\d+x\d+$')  # Check for valid format
celeb = celeb[valid_resolutions]  # Keep only valid rows

# Separate resolution into width and height
celeb[['width', 'height']] = celeb['resolution'].str.split('x', expand=True).astype(int)

# Descriptive statistics for width and height
width_stats = celeb['width'].describe().rename("Width")
height_stats = celeb['height'].describe().rename("Height")

# Combine descriptive statistics into one DataFrame
stats_df = pd.concat([width_stats, height_stats], axis=1)

# Print the statistics
print(stats_df)

# Plot the descriptive statistics as a table
fig, ax = plt.subplots(figsize=(6, 2))
ax.axis('tight')
ax.axis('off')
table = ax.table(cellText=stats_df.values, colLabels=stats_df.columns, rowLabels=stats_df.index, loc='center')
table.auto_set_font_size(False)
table.set_fontsize(10)
plt.title("Descriptive Statistics for Width and Height", fontsize=12, weight='bold')
plt.show()
```

### Repeat process for incorrect_guess_data.csv

### Qualitative variables -- add bar graphs for race and gender

```python

# Load the dataset
celeb_images = pd.read_csv("/workspace/DS4002Project3/DATA/image_data/celeb_image_data/celeb_image_data.csv")

# Set the color and border style for the bars
bar_color = 'gray'
bar_edge_color = 'black'

# RACE DISTRIBUTION
if 'race' in celeb_images.columns:
    # Group "White-Latina" and "White-Latino" with "White"
    celeb_images['race_grouped'] = celeb_images['race'].replace(
        {'White-Latina': 'White', 'White-Latino': 'White'}
    )

    # Count the occurrences of each race
    race_counts = celeb_images['race_grouped'].value_counts()

    # Plot the bar graph
    plt.figure(figsize=(8, 5))
    sns.barplot(x=race_counts.index, y=race_counts.values, color=bar_color, edgecolor=bar_edge_color)
    plt.title("Race Distribution in Celeb Images Dataset", fontsize=14, weight='bold')
    plt.xlabel("Race", fontsize=12)
    plt.ylabel("Count", fontsize=12)
    plt.xticks(rotation=45)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

# GENDER DISTRIBUTION
if 'gender' in celeb_images.columns:
    # Count the occurrences of each gender
    gender_counts = celeb_images['gender'].value_counts()

    # Plot the bar graph
    plt.figure(figsize=(8, 5))
    sns.barplot(x=gender_counts.index, y=gender_counts.values, color=bar_color, edgecolor=bar_edge_color)
    plt.title("Gender Distribution in Celeb Images Dataset", fontsize=14, weight='bold')
    plt.xlabel("Gender", fontsize=12)
    plt.ylabel("Count", fontsize=12)
    plt.xticks(rotation=0)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()
```
